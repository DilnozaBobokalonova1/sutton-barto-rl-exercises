```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def softmax(x):
    a = np.exp(x)
    b = np.sum(a)
    return a / b
```

```python
xs = np.arange(-3, 0, 0.3)
```

```python
df = pd.DataFrame(xs, columns=['logits'])
```

```python
df['softmax'] = softmax(xs)
```

```python
T = 3
df['softmax_scaled_up'] = softmax(xs * T)
df['softmax_scaled_down'] = softmax(xs / T)
```

```python
df.drop('logits', axis=1).plot.bar()
```

Using T, you could calibrate softmax in multiple ways

```python
df[['logits', 'softmax']].plot.bar()
```

```python
xs = np.arange(-10, 10, 0.1)
ys = softmax(xs) - 0.5

plt.plot(xs, ys)
```

```python
xs1 = np.arange(-10 + 0.1, 0.1, 0.1)
xs2 = np.arange(0, 10, 0.1)
ys1 = softmax(xs1) - 0.5
ys2 = - (softmax(xs2) - 0.5)[::-1]
```

```python
plt.plot(xs1, ys1)
plt.plot(xs1, ys2)
```

Perfect overlapping shows the symmetry
